```python
#Transpose of a Square Matrix
n = int(input("Enter size of matrix: "))
matrix = []
print("Enter matrix elements:")

for i in range(n):
    row = []
    for j in range(n):
        row.append(int(input()))
    matrix.append(row)
transpose = []
for i in range(n):
    row = []
    for j in range(n):
        row.append(matrix[j][i])
    transpose.append(row)
print("Transpose Matrix:")
for i in transpose:
    print(i)
```

```python
#Place Even Elements at Even Indexes and Odd Elements at Odd Indexes
n = int(input("Enter number of elements: "))
arr = []
for i in range(n):
    arr.append(int(input()))
result = [0] * n
even = 0
odd = 1
for i in range(n):
    if arr[i] % 2 == 0:
        result[even] = arr[i]
        even += 2
    else:
        result[odd] = arr[i]
        odd += 2
print(result)
```

```python
#Check if One Array is a Subsequence of Another
n1 = int(input("Enter size of first array: "))
arr1 = []
for i in range(n1):
    arr1.append(int(input()))
n2 = int(input("Enter size of second array: "))
arr2 = []
for i in range(n2):
    arr2.append(int(input()))
i = 0
for j in range(n2):
    if i < n1 and arr1[i] == arr2[j]:
        i += 1
if i == n1:
    print("Subsequence")
else:
    print("Not Subsequence")
```

```python
#Unzip nested dicts into expanded dicts
data = {
    "name": "John",
    "address": {
        "city": "Hyderabad",
        "state": "Telangana"}}

result = {}
for key, value in data.items():
    if isinstance(value, dict):
        for k, v in value.items():
            result[f"{key}_{k}"] = v
    else:
        result[key] = value

print(result)
```

```python
#Make a given string a valid Python variable name
import re
s = "123 my-variable"
s = re.sub(r'\W', '_', s)
if s[0].isdigit():
    s = "_" + s
print(s)
```

```python
#Return duplicate elements in sorted order
lst = [5, 2, 7, 5, 2, 8, 1]
duplicates = sorted(set(x for x in lst if lst.count(x) > 1))
print(duplicates)
```

```python
#Sort a list without using sort()
arr = [5, 2, 8, 1, 4]
n = len(arr)
for i in range(n):
    for j in range(n - i - 1):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]
print(arr)
```

```python
#Swap two numbers
a = 10
b = 20
a, b = b, a
print(a, b)
```

```python
#Find the maximum number in a list
arr = [12, 45, 6, 89, 32]
maximum = arr[0]
for num in arr:
    if num > maximum:
        maximum = num

print(maximum)
```

```python
#Find the unique number in a list
arr = [2, 3, 2, 4, 4]
for num in arr:
    if arr.count(num) == 1:
        print(num)
```

```python
#Remove duplicates from a list
arr = [1, 2, 2, 3, 4, 3, 5]
result = []
for x in arr:
    if x not in result:
        result.append(x)
print(result)
```
